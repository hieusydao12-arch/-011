# 11주차 실습: 난장판 폴더 확장자별 자동 청소하기

이 실습 노트북은 파일 확장자를 기준으로 자동 폴더 생성 및 이동을 수행하는 파이썬 스크립트를 작성하는 공간입니다. **코드를 외워서 직접 짜지 마세요.** 아래 단계에 따라 AI(Gemini)에게 요구하고 코드를 가져와서 에러를 해결해 보세요.

## [1단계] 난장판(더미) 파일 준비하기
먼저, 청소할 가짜 쓰레기 파일들이 필요합니다. 코랩 환경 내에 임의의 파일 100개를 무작위로 생성하는 코드를 제미니에게 물어보고, 아래 셀에 붙여넣어 실행하세요.

* 💡 **프롬프트 팁**: "코랩 환경의 'downloads'라는 폴더 안에 임의의 가짜 파일 100개를 만들어주는 파이썬 코드를 짜줘. 확장자는 pdf, docx, jpg, png, txt, xlsx가 섞여 있게 해줘."

```python
import os
import random

# Tạo thư mục 'downloads' nếu chưa có
folder_name = 'downloads'
if not os.path.exists(folder_name):
    os.makedirs(folder_name)

# Danh sách các phần mở rộng tệp giả định
extensions = ['jpg', 'png', 'pdf', 'docx', 'xlsx', 'txt']

# Tạo 100 tệp tin ngẫu nhiên
for i in range(100):
    ext = random.choice(extensions)
    file_name = f"test_file_{i}.{ext}"
    file_path = os.path.join(folder_name, file_name)

    # Tạo tệp rỗng
    with open(file_path, 'w') as f:
        f.write("Đây là nội dung tệp giả lập.")

print(f"Đã tạo xong 100 tệp tin rác trong thư mục '{folder_name}'.")
```

## [2단계] 확장자별 자동 분류 스크립트 작성
이제 위에서 만든 수많은 파일들을 확장자별로 새 폴더를 만들어 깔끔하게 이동시켜주는 분류 코드를 제미니에게 요청하고 아래 셀에 붙여넣어 실행하세요.

* 💡 **프롬프트 팁**: "'downloads' 폴더 안에 있는 파일들을 확장자별로 확인해서, 확장자 이름과 똑같은 새 폴더를 만들고 그 안으로 파일들을 모두 이동시켜주는 파이썬 코드를 만들어 줘. 이미 폴더가 있으면 만들지 말고 그냥 이동시켜 줘."

```python
import os
import shutil

base_directory = 'downloads' # Đổi tên biến để rõ ràng hơn

# Sử dụng os.scandir để duyệt thư mục hiệu quả hơn
with os.scandir(base_directory) as entries:
    for entry in entries:
        if entry.is_file(): # Kiểm tra trực tiếp xem có phải là tệp không
            file_name = entry.name
            file_path = entry.path

            if '.' in file_name:
                extension = file_name.split('.')[-1]
            else:
                extension = 'no_extension'

            destination_folder = os.path.join(base_directory, extension)

            # Tạo thư mục đích nếu chưa có. exist_ok=True ngăn lỗi nếu thư mục đã tồn tại.
            os.makedirs(destination_folder, exist_ok=True)

            shutil.move(file_path, os.path.join(destination_folder, file_name))

print("Đã xong (tối ưu hóa)")
```

## [4단계] 결과 확인: Liệt kê các tệp đã được sắp xếp

Bây giờ, hãy kiểm tra xem các tệp đã được di chuyển thành công vào các thư mục con theo phần mở rộng của chúng hay chưa. Mã dưới đây sẽ liệt kê cấu trúc thư mục và các tệp bên trong thư mục 'downloads'.

```python
import os

base_directory = 'downloads'

print(f"Cấu trúc thư mục '{base_directory}':")

# Duyệt qua thư mục gốc và tất cả các thư mục con
for root, dirs, files in os.walk(base_directory):
    level = root.replace(base_directory, '').count(os.sep)
    indent = ' ' * 4 * (level)
    print(f'{indent}{os.path.basename(root)}/')
    subindent = ' ' * 4 * (level + 1)
    for f in files:
        print(f'{subindent}{f}')
```
